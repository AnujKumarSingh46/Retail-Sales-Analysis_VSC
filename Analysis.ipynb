```python
import pandas as pd
import plotly.express as px
import os
from itertools import combinations
from collections import Counter
```

```python
file_names=os.listdir('SalesData')

edf= pd.DataFrame()

for file in file_names:
    df= pd.read_csv(fr'SalesData\{file}')
    edf= pd.concat([edf,df])

edf
```

```python
edf.to_csv('Sales_2019.csv',index=False)
```

```python
edf= edf.dropna(how='all')
edf
```

```python
edf.info()
```

```python
edf= edf.loc[edf['Quantity Ordered']!='Quantity Ordered']
edf
```

```python
edf['Quantity Ordered']= edf['Quantity Ordered'].astype(int)
edf['Price Each']= edf['Price Each'].astype(float)
```

```python
edf.info()
```

```python
edf['Order Date']=pd.to_datetime(edf['Order Date'])
```

```python
edf.info()
```

```python
edf['Year']=edf['Order Date'].dt.year
edf['Month Num']=edf['Order Date'].dt.month
edf['Month']=edf['Order Date'].dt.month_name()
edf['Day Num']= edf['Order Date'].dt.day
edf['Day']=edf['Order Date'].dt.day_name()
edf
```

```python
edf['Sale']= edf['Quantity Ordered'] * edf['Price Each']
edf
```

```python
ndf= edf.groupby(['Month Num','Month']).agg(
    Total_sale=('Sale','sum')
).reset_index()
ndf['Total_sale']=ndf['Total_sale'].round()
ndf
```

```python
fig= px.bar(data_frame=ndf,
            x='Month',
            y='Total_sale',
            text='Total_sale',color='Month')
fig.show()
```

```python
edf
```

```python
edf['City']=edf['Purchase Address'].str.split(',',expand=True)[1]
edf
```

```python
ndf= edf.groupby('City').agg(
    Total_nop=('Quantity Ordered','sum')
).reset_index()
ndf= ndf.sort_values('Total_nop',ascending=False)
ndf
```

```python
fig= px.bar(data_frame=ndf,
            x='Total_nop',
            y='City',
            text='Total_nop',color='City')
fig.show()
```

```python
fig=px.pie(data_frame=ndf,
           names='City',
           values='Total_nop',hole=0.4)
fig.show()
```

```python
edf['Hour']= edf['Order Date'].dt.hour
edf
```

```python
fig=px.histogram(data_frame=edf,
                 x='Hour',color='Product',barmode='overlay')
fig.show()
```

```python
dup_df=edf.loc[edf['Order ID'].duplicated(keep=False)]
dup_df
```

```python
dup_df['Product']=','+ dup_df['Product']
dup_df=dup_df.groupby('Order ID').agg(
    Group=('Product','sum')
)
dup_df
```

```python
dup_df['Group']=dup_df['Group'].str.strip(',')
dup_df=dup_df.reset_index()
dup_df
```

```python
dup_df.to_csv('Duplicate Orders.csv')
```

```python
dup_df['Group']=dup_df['Group'].str.split(',')
dup_df
```

```python
list= list(dup_df['Group'])
list
```

```python
count= Counter()
for i in list:
    count.update(Counter(combinations(i,2)))
combination_counts=pd.DataFrame(count.items(),columns=['Combinations','Count'])
combination_counts= combination_counts.sort_values('Count',ascending=False)
combination_counts
```

```python
combination_counts.to_csv('Combination Count Sorted.csv')
```

```python
data=combination_counts.head(10)
data
```

```python
fig=px.pie(data_frame=data,
           names='Combinations',
           values='Count',
           title='Top 10 Combinations of Products Sold')
fig.update_traces(pull=[0.1 if i == 0 else 0 for i in range(len(data))])  # Explode only the first slice

fig.show()
```
